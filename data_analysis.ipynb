```python
import pandas as pd
data1 = pd.read_excel('商品销售数据.xlsx',sheet_name='信息表')
data1
```

```python
data2 = pd.read_excel('商品销售数据.xlsx',sheet_name='销售数据表')
data2
```

```python
data_all = pd.merge(data1,data2,on="商品编号",how='right')
data_all
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


plt.rcParams["font.family"] = ["SimHei", "WenQuanYi Micro Hei", "Heiti TC"]
plt.rcParams["axes.unicode_minus"] = False  # 解决负号显示问题

# 计算销售金额
data_all['销售金额'] = data_all['商品销售价'] * data_all['订单数量']

# 从订单日期提取月份信息
data_all['月份'] = pd.to_datetime(data_all['订单日期']).dt.month

# 按月份统计销售金额总和
monthly_sales = data_all.groupby('月份')['销售金额'].sum().reset_index()

# 打印结果
print("按月份统计的销售金额总和：")
print(monthly_sales)

# 将统计结果添加到原数据中（可选）
# 这里将每月总销售金额合并回原数据，便于后续分析
data_all = pd.merge(data_all, monthly_sales, on='月份', suffixes=('', '_月总计'))

# 可视化：绘制月度销售金额柱状图
plt.figure(figsize=(10, 6))
sns.barplot(x='月份', y='销售金额_月总计', data=data_all.drop_duplicates('月份'))
plt.title('月度销售金额统计')
plt.xlabel('月份')
plt.ylabel('销售金额总和')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('月度销售金额统计.png')
plt.show()

# 保存结果到新的Excel文件
data_all.to_excel('商品销售数据分析结果.xlsx', index=False)
print("分析结果已保存到 '商品销售数据分析结果.xlsx'")
```

```python



# ① 提取“月份”和“销售金额”列数据作为 x 和 y 坐标值
x = monthly_sales['月份']
y = monthly_sales['销售金额']

# ② 使用 plot() 绘制折线图，对比商品每月销售金额
plt.plot(x, y)

# ③ 设置图表和 x 轴、y 轴标题
plt.title('商品每月销售金额折线图')
plt.xlabel('月份')
plt.ylabel('销售金额')

# ④ 为 x 轴和 y 轴设置新刻度（x 轴设置为 1 - 12 ，y 轴根据数据合理设置，这里示例简单设置）
plt.xticks(range(1, 13))  # x 轴刻度设为 1 到 12
# y 轴刻度可根据数据范围灵活调整，比如根据最大值、最小值做均分，这里示例简单设置
plt.yticks(range(200000, int(y.max()) + 200000, 100000))  

# 显示图表
plt.show()

# 若需要保存处理后的数据到新 Excel（可选，和之前保存逻辑一致）
# data2.to_excel('商品销售数据分析结果.xlsx', index=False)
```

```python


# ① 按照“月份”和“商品大类”对销售金额进行分组汇总
grouped = data_all.groupby(['月份', '商品大类'])['销售金额'].sum().reset_index()

# ② 从分组汇总结果中筛选出各个商品大类的数据（这里以示例方式，可根据实际需求遍历或按需筛选）
# 假设要获取所有不同商品大类，分别处理
categories = grouped['商品大类'].unique()

# ③ 使用 bar() 绘制柱状图，对比不同商品大类的月度销售额
for cat in categories:
    cat_data = grouped[grouped['商品大类'] == cat]
    plt.bar(cat_data['月份'], cat_data['销售金额'], label=cat, alpha=0.6)

plt.title('不同商品大类月度销售额对比')
plt.xlabel('月份')
plt.ylabel('销售金额')
plt.xticks(range(1, 13))  # 设置 x 轴刻度为 1 - 12 月
plt.legend()  # 显示图例区分不同商品大类
plt.show()
```
